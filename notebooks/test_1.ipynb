```python
from music21 import converter, note, chord

# Load the MIDI file
midi_file_path = "/import/c4dm-datasets/maestro-v3.0.0/2008/MIDI-Unprocessed_02_R1_2008_01-05_ORIG_MID--AUDIO_02_R1_2008_wav--1.midi"
midi = converter.parse(midi_file_path)

total_notes = 0
# Iterate over all the parts
for part in midi.parts:
    # Iterate over all the notes and chords in the part
    for element in part.flat.notes:
        if isinstance(element, note.Note):
            # Print the pitch, duration, onset, offset, and bar info of the note
            print('Note: {}, Duration: {}, Onset: {}, Offset: {}, Bar: {}'.format(element.pitch, element.duration, element.offset, element.offset + element.duration.quarterLength, element.measureNumber))
            total_notes += 1
        elif isinstance(element, chord.Chord):
            # Print the pitches, duration, onset, offset, and bar info of the chord
            print('Chord: {}, Duration: {}, Onset: {}, Offset: {}, Bar: {}'.format(' '.join(str(n) for n in element.pitches), element.duration, element.offset, element.offset + element.duration.quarterLength, element.measureNumber))
            for pitch in element.pitches:
                total_notes += 1

print(total_notes)

# # Convert the MIDI part back to a MIDI file
# output_midi_file_path = "music_21_test.mid"
# midi.write('midi', fp=output_midi_file_path)
```
